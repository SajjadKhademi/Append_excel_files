```python
protein_dict = {"name":"protein" , "data":protein }
bakery_dict = {"name":"bakery" , "data":bakery }
cafe_dict = {"name":"cafe" , "data":cafe }
confectionery_dict = {"name":"confectionery" , "data":confectionery}
fruit_dict = {"name":"fruit" , "data":fruit}
juice_dict = {"name":"juice" , "data":juice}
restaurant_dict = {"name":"restaurant" , "data":restaurant}
```

```python
services = [protein_dict, bakery_dict, cafe_dict, confectionery_dict, fruit_dict ,juice_dict ,restaurant_dict]
services
```

```python
df = pd.DataFrame()
for serv in services:
    serv['data']['Service'] = serv['name']
    df = pd.concat([df,serv['data']], ignore_index=True)
df
```

```python
df['Metric'].unique()
Used_metrics = ['01- Total Orders','07- NMV','10- Acq. Voucher Cost','11- Ret. Voucher Cost','12- Other Cities Voucher Cost',\
                '13- Non-Voucher DB Cost','14- SMS & Other Non-DB Costs','15- ATL Cost', '16- Competition Cost']
```
